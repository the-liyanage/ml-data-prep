### 5. Calculate Missing Values as Percentage

- `df.isnull()` detects missing values, returning boolean DataFrame

- Sum missing across columbs, normalize by total rows and multiply by 100


```python
total_null_values = df.isnull().sum()
print(total_null_values)
```

```python
avg_total_null_values = (df.isnull().sum() /df.shape[0])*100
print(avg_total_null_values)
```

```python
round_nan_values = round(avg_total_null_values, 2)
print(round_nan_values)
```

### 6. Drop Irrelevant or Data-Heavy Missing Columns

- `df.drop(columns =[])` --> Drops specified columns from the DataFrame.

- `df.dropna(subset=[])`  --> Removes rows where specified columns have missing values.

- `fillna()` ---> Fills missing values with specified values(mean).

```python
#Dropping columns 

df1 = df.drop(columns=["Name", "Ticket", "Cabin"])
df1.head()
```

```python
#Dropping columns with missing values 
df1.dropna(subset=["Embarked"], inplace = True)
df1.head()
df1.isnull().sum()
```

```python
#Fills missing values with specified values 
df1["Age"].fillna(df1["Age"].mean(), inplace = True)
df1.head()
df1.isnull().sum()
```

### 7. Detect Outliers with Box Plot

- `matplotlib.pyplot.boxplot()` --> Displays distribtion of data, highlighting median, quartiles and outliers.

- `plt.show()` --> Renders the plot

```python
import matplotlib.pyplot as plt

plt.boxplot(df1["Age"], vert = False)
plt.ylabel("Variable")
plt.xlabel("Age")
plt.title("Box plt")
plt.show()
```

### 8. Calculate Outlier Boundaries and Remove Them

- Calculate `mean` and `standard deviation (std)` using  `df["Age"}.mean` and `df["Age"].std()`

- Define bounds as mean  ±2 * std for outlier detection.

- Filter DataFrame rows within bounds using Boolean indexing.

```python
mean = df1["Age"].mean()
print(mean)
print()

std = df1["Age"].std()
print(std)
```

```python
lower_bound = mean - 2 * std
upper_bound = mean + 2 * std

print(f"Lower bound:{lower_bound:.2f}, Upper bound:{upper_bound:.2f}")
```

```python
df2 = df1[(df1["Age"]>= lower_bound) & (df1["Age"] <= upper_bound)]
df2.head()
```

### 9. Impute Missing Data Again if Any

- `fillna()` applied again on filtered data to handle any remaining missing value.

```python
df3 = df2.fillna(df2["Age"].mean())
df3.isnull().sum()
```

### 10. Recalculate Outlier Bounds and Remove Outliers from the Updated Data

```python
mean = df3["Age"].mean()
print(f"{mean}\n")
std = df3["Age"].std()
print(f"{std}\n")
```

```python
lower_bound = mean - 2 * std
upper_bound = mean + 2 * std

print(f"Lower bound:{lower_bound:.2f}, Upper bound:{upper_bound:.2f}")
```

```python
df4 = df3[(df3["Age"]>= lower_bound) & (df3["Age"] <= upper_bound)]
df4.head()
```

### 12. Data Formatting 

- Data formatting involves converting the data into a standard format or structure that can be easily processed by the algorithms or models.

- Using `Scaling` and `Normalization.`

- `Scaling` --> Transforming the values of features to a specific range.
                It maintains the shape of the original distribution while changing                 the scale.
                It is useful when features have different scales and certain                       algortihms are sensitive to the magnitude of the features


1. `Min - Max Scaling` --> rescale the values to a specified range, typically btetween 0 - 1.
   It preserves the original distribition and ensures that the minimum value to 0 and the maximum value maps to 1


```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range = (0,1))

num_col_ = [col for col in X.columns if X[col].dtype !='object']
x1 = X
x1[num_col_] = scaler.fit_transform(x1[num_col_])
x1.head()
```

2. `Standardization (Z- score scaling)` --> Standardization transforms the values                             to have a mean of 0 and a standard deviation of 1.<br>
                         - Standardization transforms the values to have a mean of 0 and a standard deviation of 1.<br>
                         - Standardization makes the data more suitable for <br> algorithms that assume a Gaussian distribution or require features to have zero<br> mean and unit variance. <br><br>



   `Z = (X - μ) / σ`<br>
   `Where`,<br>

    `X = Data 
     μ = Mean value of X 
     σ = Standard deviation of X `
