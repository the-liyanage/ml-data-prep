### Using Numpy Library

```python
import numpy as np
```

```python
x = [215, 325, 185, 332, 406, 522, 412,
     614, 544, 421, 445, 408],
y = [14.2, 16.4, 11.9, 15.2, 18.5, 22.1, 
     19.4, 25.1, 23.4, 18.1, 22.6, 17.2]

matrix = np.corrcoef(x, y)
print(matrix)
```

### Using Pandas Library 

```python
import pandas as pd
```

```python
data = {
    'x': [45, 37, 42, 35, 39],
    'y': [38, 31, 26, 28, 33],
    'z': [10, 15, 17, 21, 12]
}

dataframe = pd.DataFrame(data, columns = ['x', 'y', 'z'])
print(f"DataFrame is \nnn{dataframe}\n\n")

#Creating a Correlation Matrix
matrix = dataframe.corr()
print(matrix)
```

### Using Matplotlib and Seaborn for Visualization 


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
matrix = dataframe.corr()

plt.figure(figsize = (8,6))
sns.heatmap(matrix, annot = True, cmap = "coolwarm", fmt = ".2f", linewidth = 0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
#Importing the dataset from sklearn

from sklearn import datasets
```

### Understanding Correlation Values
 - No Correlation: A correlation value of 0 means no linear relationship between
the variables. As one changes, the other does not follow any predictable pattern.
    
- Positive Correlation: A value closer to +1 indicates a direct relationship 
as one variable increases, the other also increases. Example: height and weight.

- Negative Correlation: A value closer to -1 indicates an inverse relationship 
as one variable increases, the other decreases. Example: speed and travel time.
