### Detecting Multicollinearity with VIF - Python

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("bmi.csv.xls")
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

#Converting Categorical varibles into numeric form
df["Gender"] = df["Gender"].map({"Male":0, "Female":1})
X = df[["Gender", "Height", "Weight"]]

X.isna().sum()
```

```python
np.isinf(X).sum()
```

```python
X.isnull().sum()
```

```python
X = X.dropna()
X = X.fillna(X.mean())
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

#Converting Categorical varibles into numeric form
df["Gender"] = df["Gender"].map({"Male":0, "Female":1})
X = df[["Gender", "Height", "Weight"]]


#Passing each feature index to the variance_inflation_factor() to calculate VIF
vif_df = pd.DataFrame()
vif_df["feature"] = X.columns

vif_df["VIF"] = [variance_inflation_factor(X.values,i) for i in range(len(X.columns))]

print(vif_df)
```
